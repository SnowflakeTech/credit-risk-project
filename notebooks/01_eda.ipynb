```python
target = "default.payment.next.month"

df[target].value_counts()
```

```python
df[target].value_counts(normalize=True)
```

```python
sns.countplot(x=df[target])
plt.title("Distribution of Default vs Non-Default")
plt.show()
```

```python
sns.barplot(x="SEX", y=target, data=df)
plt.title("Default Rate by Gender")
plt.show()
```

```python
sns.barplot(x="EDUCATION", y=target, data=df)
plt.title("Default Rate by Education")
plt.show()
```

```python
sns.barplot(x="MARRIAGE", y=target, data=df)
plt.title("Default Rate by Marriage Status")
plt.show()
```

```python
sns.histplot(df["AGE"], bins=30, kde=True)
plt.title("Age Distribution")
plt.show()
```

```python
df["AGE_GROUP"] = pd.cut(df["AGE"], bins=[20,30,40,50,60,80])
```

```python
sns.barplot(x="AGE_GROUP", y=target, data=df)
plt.xticks(rotation=45)
plt.title("Default Rate by Age Group")
plt.show()
```

```python
sns.histplot(df["LIMIT_BAL"], bins=50)
plt.title("Credit Limit Distribution")
plt.show()
```

```python
pay_cols = ["PAY_0","PAY_2","PAY_3","PAY_4","PAY_5","PAY_6"]

df[pay_cols].describe()
```

```python
df[pay_cols].mean().plot(kind="bar")
plt.title("Average Payment Delay by Month")
plt.show()
```

```python
pay_amt_cols = ["PAY_AMT1","PAY_AMT2","PAY_AMT3","PAY_AMT4","PAY_AMT5","PAY_AMT6"]

df[pay_amt_cols].mean().plot()
plt.title("Average Payment Amount Trend")
plt.show()
```

```python
# chỉ lấy các cột số
numeric_cols = df.select_dtypes(include='number')  # int, float

plt.figure(figsize=(12,8))
sns.heatmap(numeric_cols.corr(), cmap="coolwarm", annot=True, fmt=".2f")
plt.title("Correlation Matrix")
plt.show()
```
